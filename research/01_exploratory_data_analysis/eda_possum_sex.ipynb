```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Explore feature Sex of possum dataset

```python
possum = pd.read_csv('possum.csv')
```

```python
print(possum.head())
```

```python
print(possum.describe())
```

```python
print(possum['site'].nunique())
```

```python
fig, axs = plt.subplots(5, 2, figsize=(12, 35))
sns.boxplot(data=possum, x='sex', y='hdlngth', ax=axs[0, 0])
sns.boxplot(data=possum, x='sex', y='skullw', ax=axs[0, 1])
sns.boxplot(data=possum, x='sex', y='totlngth', ax=axs[1, 0])
sns.boxplot(data=possum, x='sex', y='taill', ax=axs[1, 1])
sns.boxplot(data=possum, x='sex', y='footlgth', ax=axs[2, 0])
sns.boxplot(data=possum, x='sex', y='earconch', ax=axs[2, 1])
sns.boxplot(data=possum, x='sex', y='eye', ax=axs[3, 0])
sns.boxplot(data=possum, x='sex', y='chest', ax=axs[3, 1])
sns.boxplot(data=possum, x='sex', y='belly', ax=axs[4, 0])
plt.show()
```

## => Insight: most correlated with sex seem to be earconch and footlgth
