```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Explore feature Site of possum dataset

```python
possum = pd.read_csv('possum.csv')
```

```python
print(possum.head())
```

```python
print(possum.describe())
```

```python
print(possum['site'].nunique())
```

```python
fig, axs = plt.subplots(5, 2, figsize=(15, 30))
sns.boxplot(data=possum, x='site', y='hdlngth', ax=axs[0, 0])
sns.boxplot(data=possum, x='site', y='skullw', ax=axs[0, 1])
sns.boxplot(data=possum, x='site', y='totlngth', ax=axs[1, 0])
sns.boxplot(data=possum, x='site', y='taill', ax=axs[1, 1])
sns.boxplot(data=possum, x='site', y='footlgth', ax=axs[2, 0])
sns.boxplot(data=possum, x='site', y='earconch', ax=axs[2, 1])
sns.boxplot(data=possum, x='site', y='eye', ax=axs[3, 0])
sns.boxplot(data=possum, x='site', y='chest', ax=axs[3, 1])
sns.boxplot(data=possum, x='site', y='belly', ax=axs[4, 0])
plt.show()
```

## => Insight: We have great differences in all features!
