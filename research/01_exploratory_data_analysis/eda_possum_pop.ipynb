```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Explore feature Pop of possum dataset

```python
possum = pd.read_csv('possum.csv')
```

```python
print(possum.head())
```

```python
print(possum.describe())
```

```python
print(possum['Pop'].nunique())
```

```python
fig, axs = plt.subplots(5, 2, figsize=(10, 35))
sns.boxplot(data=possum, x='Pop', y='hdlngth', ax=axs[0, 0])
sns.boxplot(data=possum, x='Pop', y='skullw', ax=axs[0, 1])
sns.boxplot(data=possum, x='Pop', y='totlngth', ax=axs[1, 0])
sns.boxplot(data=possum, x='Pop', y='taill', ax=axs[1, 1])
sns.boxplot(data=possum, x='Pop', y='footlgth', ax=axs[2, 0])
sns.boxplot(data=possum, x='Pop', y='earconch', ax=axs[2, 1])
sns.boxplot(data=possum, x='Pop', y='eye', ax=axs[3, 0])
sns.boxplot(data=possum, x='Pop', y='chest', ax=axs[3, 1])
sns.boxplot(data=possum, x='Pop', y='belly', ax=axs[4, 0])
plt.show()
```

## => Insight: We have great differences especially in footlgth, earconch, totlgth, tail and chest!
